```python
# import requests as req
import pandas as pd
import time
from bs4 import BeautifulSoup as bs

# Service와 CDM은 세트
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Keys : 키 사용, By : 선택자
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 크롬 드라이버 설치
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# os로 경로 설정 + 파일저장
import os
from urllib.request import urlretrieve
```

```python
url = 'https://corners.gmarket.co.kr/Bestsellers';
driver.get(url)
time.sleep(1)

img = driver.find_elements(By.CSS_SELECTOR, 'img.lazy')

title_list = []
price_list = []
where_list = []

for i in range(10):
    img = driver.find_elements(By.CSS_SELECTOR, 'img.lazy')
    img[i].click()
    time.sleep(0.25)
    
    soup = bs(driver.page_source, 'lxml')
    span = soup.select('span.sp_vipgroup')
    for i in span:
        i.extract()
    title = soup.select_one('h1.itemtit').text
    price = soup.select_one('strong.price_real').text
    where = soup.select_one('li.list-item-origin.uxeslide_item div.box__information').text.strip()
    
    title_list.append(title)
    price_list.append(price)
    where_list.append(where)
    
    driver.back()
    time.sleep(0.25)
```

```python
origin_list = []
for i in range(len(where_list)):
    origin_list = where_list[i][3 : ]

dic = {'제품명': title_list, '가격': price_list, '원산지': origin_list}
df = pd.DataFrame(dic)
df
```
